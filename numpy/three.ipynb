```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## data structure:[restaurant_id, 2021,2022,2023,2024]
sales_data=np.array([
    [1, 130000, 1400000, 258000, 457000],
    [2, 350000, 1400000, 310000, 340000],
    [3, 130000, 3470000, 350000, 230000],
    [4, 450000, 2300000, 780000, 234000]
])

print('slaes data shape',sales_data.shape)
print('sample data  fpr first three restaurant \n',sales_data[0:3])
```

```python
##total sales per year
year_sales=np.sum(sales_data[:,1:],axis=0)
print(year_sales)
print(np.sum(year_sales))

##minimum sales per restaurant
min_sales=np.min(sales_data[:,1:],axis=1)
print('Minimum sales: ',min_sales)

##maximum sales per year
max_sales=np.max(sales_data[:,1:],axis=0)
print('maximum sales per year',max_sales)

##Average sales per restaurant
avg_sales=np.mean(sales_data[:,1:],axis=0)
print(avg_sales)


 
```

```python
##cumulataive sales
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

## Plotting result

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title('Average Sales per year')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

## Vector

```python
vector1=np.array([1,2,3])
vector2=np.array([3,4,5])
print('addition pf two  vectors is: ',vector1+vector2)
print('Vector multiplication : ', vector1*vector2)
print('Dot product of the vector: ',np.dot(vector1,vector2))
angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)


##vectorized operations
restaurant=np.array(['1star','2star','3star'])
upper=np.vectorize(str.upper)
print('Vectorized Upper',upper(restaurant))
```
