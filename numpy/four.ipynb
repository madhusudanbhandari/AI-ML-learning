## Logo

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
logo = mpimg.imread('foto.png')   # put your actual image file name here

# Display the shape to confirm itâ€™s an array
print("Shape:", logo.shape)

```

```python
np.save('numpy-logo.npy', logo)
print("Saved successfully as numpy-logo.npy")
```

```python
logo=np.load('numpy-logo.npy')
plt.imshow(logo)
plt.axis('off')  # Hide the axis
plt.title('Numpy logo')
plt.show()

```

```python
dark=1-logo
plt.imshow(dark)
plt.axis('off')  # Hide the axis
plt.title('Numpy logo')
plt.show()
```
