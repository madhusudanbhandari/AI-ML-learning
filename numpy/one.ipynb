```python
import numpy as np
```

```python
#printing the arrays from list
arr_1d=np.array([1,2,3,4])
print('1d array is:',arr_1d)

arr_2d=np.array([[1,2,3],[4,5,7]])
print('2d array is:',arr_2d)
```

```python
py_list=[1,2,4]
print('python list multiplication',py_list*2)

np_array=np.array([1,2,3])
print('Python array multiplication',np_array*2)

import time
start=time.time()
py_list=[i*2 for i in range(100)]
print('time for list operation is: ',time.time()-start)

start=time.time()
np_array=np.arange(100000)*2
print('time for numpy operation is: ',time.time()-start)
```

```python
##creating array from scratch
zeroes=np.zeros((2,3))
print('zeroes array \n',zeroes)

ones=np.ones((2,3))
print('ones array \n',ones)

random=np.random.random((2,3))
print('random array \n',random)

sequence=np.arange(0,10,2)
print('sequence array \n',sequence)
```

```python
###Vector, Matrix and Tensor
vector=np.array([1,2])
print('Vector:\n',vector)

matrix=np.array([[1,2],[3,4]])
print('Matrix:\n',matrix)

tensor=np.array([[[1,2],[4,3]],[[3,4],[4,5]]])
print('Tensor: \n',tensor)
```

```python
###Array Properties
arr=np.array([[1,2,3],
           [4,5,6]])
print('shape:',arr.shape)
print('size',arr.size)
print('dimension:',arr.ndim)
print('type:',arr.dtype)
```

```python
##Arrat Reshaping
arr=np.arange(1,13)
print('array:',arr)

reshaped=arr.reshape(3,4)
print('Reshaped Array:',reshaped)

flattened=reshaped.flatten()
print('Flattened Array:',flattened)

raveled=reshaped.ravel()
print('Raveled Array:',raveled)

transpose=reshaped.T
print('Transpose of matrix is ',transpose)
```
