## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python

X=[1,2,3,4]
y=[5,3,7,2]
plt.plot(X,y)
plt.show()
```

```python
days=['mon','tue','wed','thurs']
sales=[23,34,12,43]
exepense=[34,56,2,22]

plt.plot(days,sales,exepense)
plt.title('Sales Data')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.legend('Sales')
plt.show()
```

```python
months=['jan','feb','mar','apr','may']
sales=[2300,4500,5500,3300,3700]

plt.plot(months,sales, color='blue', linestyle='--', linewidth=2, marker='o',label='2026 sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales Data')
plt.legend()
plt.grid()
# plt.xlim('jan','apr')
# plt.ylim(2300,3300)

plt.xticks(['jan','feb','mar','apr','may'],['January','febraury','march','april','may'])
plt.show()
```

```python
product=[1,2,3,4]
sales=[1000,400,1400,2000]

plt.bar(product,sales,color='red',label='Sales data')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Produc sales comparision')
plt.legend()
plt.show()
```

```python
regions=['North','South','East','West']
revenue=[1000,2000,5000,6000]

plt.pie(revenue,labels=regions, autopct='%1.1f%%',colors=['red','green','blue','yellow'] )
plt.title('Revenue from different regions')
plt.show()
```

```python
##histogram==based on the intervals

scores=[3,34,53,66,66,66,8,6,33,44,54,45,47,88,66,63,36]

plt.hist(scores, bins=10,color='green',edgecolor='black')
plt.xlabel('Scores')
plt.ylabel('Number of students')

plt.title('Marks distribution of the students')
plt.show()
```

## Scatter Plots

```python
##shows the relation between two data

hours_studied=[1,2,3,4,5,6,7,8]
marks=[40,45,50,55,60,65,70,75]

plt.scatter(hours_studied,marks,color='green',marker='o',label='Student Data')
plt.title('Relationship between studyhours and marks')
plt.xlabel('Study Hours')
plt.ylabel('marks')
plt.grid()
plt.show()
```
