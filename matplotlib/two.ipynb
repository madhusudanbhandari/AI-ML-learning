## Subplots

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,25,50,24]

plt.subplot(1,2,1) ##1 row, 2 column ,1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Chart')
plt.show()
```

```python
x=[1,2,3,4]
y=[10,25,50,24]

fig,ax=plt.subplots(1,2,figsize=(10,5))

ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

plt.show()
```

```python
x=[1,2,3,4]
y=[10,15,20,8]

fig, ax=plt.subplots(1,2,figsize=(10,5))

ax[0].plot(x,y)
ax[0].set_title('Line Chart')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

plt.tight_layout()
fig.suptitle('Comparision of line and bar charts')

plt.show()
```

## Save Figures

```python
x=[1,2,3,4]
y=[10,20,15,25]
plt.plot(x,y,color='blue',marker='o')
plt.title('Simple line plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.savefig('line_plot.png',dpi=300,bbox_inches='tight')
plt.show()
```
