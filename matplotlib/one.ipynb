## MATPLOTLIB


```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[20,13,26,18]
plt.plot(x,y)
plt.show()
```

```python
x=['mon','tues','wed','thur','fri']
y=[10,15,7,20,12]
plt.plot(x,y)
plt.title('Bakery Sales This week')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.show()
```

## Pyplot Functions

```python

months=[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.plot(months,sales,color='blue',linestyle='--',linewidth=2,marker='o',label='2025 Sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Data')
plt.legend(loc='lower right')  ##shows the label if it was set
plt.grid(color='grey',linestyle=':',linewidth=1)  
plt.xlim(0,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])   ##Replaces the x axis values with new values similarly yticks for y axis
plt.show()
```

## Bar charts , pie charts, histogram

```python
#Bar Chart
product=['A','B','C','D']
sales=[1000,1500,800,1300]

plt.bar(product,sales,color='red',label='Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales Data')


plt.savefig('bargraph.png',dpi=300,bbox_inches='tight')
plt.legend()
plt.show()
```

```python
##Pie chart
revenue=[1000,1500,800,1200]
regions=['north','south','east','west']
plt.pie(revenue,labels=region,autopct='%1.1f%%',colors=['red','green','yellow','grey'])
plt.title('Pie chart for revenue data')
plt.show()
```

```python
##Histogram
scores=[45,67,89,56,78,88,92,60,59,67,49,89]
plt.hist(scores,bins=5,color='green',edgecolor='black')
plt.xlabel('Score Ranges')
plt.ylabel('Number of Students')
plt.title('Score Distribution of students')
plt.show()
```

## Scatter Plot

```python
 hours=[1,2,3,4,5,6,7,8]
 score=[30,40,50,60,70,80,90,100]
 plt.scatter(hours,score,color='green',marker='o',label='Student data')
 plt.xlabel('Hours study')
 plt.ylabel('Exam Score')
 plt.title('Relationship between hours study and exam score')
 plt.legend()
 plt.grid(True)
 plt.show()
```

```python
 plt.scatter([1,2,3],[30,40,50],color='green',marker='o',label='Student data')
 plt.scatter([1,2,3],[40,50,60],color='red',marker='*',label='Student data')
 plt.xlabel('Hours study')
 plt.ylabel('Exam Score')
 plt.title('Relationship between hours study and exam score of 2 class')
 plt.legend()
 plt.grid(True)
 plt.show()
```
