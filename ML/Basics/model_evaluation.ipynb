## Confusion Metrics


```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

y_true=[1,0,0,1,1,1,0]
y_pred=[0,1,1,1,0,0,0]

print('Accuracy=',accuracy_score(y_true,y_pred))
print('Precison= ',precision_score(y_pred,y_true))
print('Recall= ',recall_score(y_pred,y_true))
print('f1_score= ',f1_score(y_pred,y_true))
```

```python
from sklearn.metrics import confusion_matrix

y_true=[1,0,1,0,0,1,0,1,1]
y_pred=[1,1,1,0,0,0,1,0,1]
cm=confusion_matrix(y_true,y_pred)
print('Confusion Matrix')
print(cm)
```

## MAE, MSE, RMSE

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error

y_true=[90,60,80,100]
y_pred=[85,70,70,95]

print('Mean squared error is',mean_squared_error(y_pred,y_true))
print('mean_absolute_error is ',mean_absolute_error(y_pred,y_true))
print('Root mean squared error is',root_mean_squared_error(y_pred,y_true))
```
