## Linear Regression

```python
#finds  a pattern in old daata

from sklearn.linear_model import LinearRegression

X=[[1],[2],[3],[4],[5]]
y=[40,50,65,75,90]

model=LinearRegression()
model.fit(X,y)

hours=float(input('Enter hours you studied? '))
predicted_marks=model.predict([[hours]])
print('You will  Score', predicted_marks,'marks')
```

```python
from sklearn.linear_model import LinearRegression

work=[[2],[3],[4],[5],[6]]
wage=[3400,5600,5300,6400,7200]

model=LinearRegression()

model.fit(work,wage)

work_hours=float(input('Enter the hours you worked'))
expected_wage=model.predict([[work_hours]])

print('You will earn',expected_wage)
```

## Classification

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

X=[[1],[2],[3],[4],[5]]
y=[0,0,1,1,1]

model=LogisticRegression()
model.fit(X,y)
hours=float(input('Enter hours you studied: '))

result=model.predict([[hours]])[0]

if result==1:
    print(f'Based on your {hours} hours of study you are likely to pass')
else:
    print('You are likely to fail')
```

## KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier

#fruit classifier
X=[[180,7],[200,7.5],[250,8],[300,8.5],[330,9],[360,9.5]]  #weight and size
y=(0,0,0,1,1,1) ##0=apple and 1=orange

model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

weight=float(input('Enter the weight in grams'))
size=float(input('Enter the size in cm'))

prediction=model.predict([[weight,size]])[0]

if prediction==0:
    print('This is likely to be apple')
else:
    print('This is likely to be orange')
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
X=[[7,2],[8,3],[9,8],[10,9]]
y=[0,0,1,1]  #0=apple ,1-orange

model=DecisionTreeClassifier()

model.fit(X,y)

size=float(input('Enter the size in cm: '))
shade=float(input('Enter the color: '))

result=model.predict([[size,shade]])[0]

if result==0:
  print('Based on given data this is apple')
else:
 print('this is an orange')
```
